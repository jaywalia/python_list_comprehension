### Basic Iteration mechanism

We generate a list of integers.

```python

LIM = HUNDRED

# initalize
integers = [0] * LIM
# print(integers)

# set values 
for i in range(LIM):
#     print(i)
    integers[i] = i

print(integers)
```

Lets separate evens out.

```python
even_integers = []
for i in range(len(integers)):
    if integers[i]%2 == 0:
        even_integers.append(integers[i])
print(even_integers)
```

Maybe we need the squares of the even integers

```python
squared_even_integers = []
for i in range(len(even_integers)):
    squared_even_integers.append(even_integers[i]**2)
print(squared_even_integers)
```

Ok, now give me the sum of the squared even integers

```python
sum_of_evens = 0
for i in range(len(squared_even_integers)):
    sum_of_evens += squared_even_integers[i]

print(f"sum_of_evens : {sum_of_evens}")

# another way to do the same
s_o_e2 = 0
for x in squared_even_integers: s_o_e2 += x
print(s_o_e2)

```

For **Cleaner code**, use list comprehension.

We can just get the odd integers

```python
odd_integers = [x for x in integers if x%2 != 0]
print(odd_integers)
```

Or we can get the squares of the odd integers, 
we are mapping a function and applying a filter at the same time.

```python
squared_odd_integers =  [x**2 for x in integers if x%2 != 0]
print(squared_odd_integers)
```

So, how about a reduce, 
normally list comprehension is to produce a list, not a single value
but reduce is available in python
and so are lambdas (recall delegates, anonymous functions from C#)

```python
import functools
sum_of_odds = functools.reduce(lambda a,b: a+b, squared_odd_integers)
print(sum_of_odds)
```

Can we apply different functions to different elements? Yes, we can!

```python
# square even and cube odds
square_even_cube_odd = [(x**2 if x%2==0 else x**3) for x in integers]
print(square_even_cube_odd)
```

How about applying many functions to all elements and then separating them in different lists? Yes, yes, we can do that too.

```python
# create two lists from one list
# https://stackoverflow.com/questions/50950690/python-unpacking-operator
squared_evens, cubed_odds = zip(*[(x**2, x**3) for x in integers])
print(squared_evens)
print(cubed_odds)
```

Hmmmm..so far so good, can we join two lists? Well, how about zipping them together.

```python
# create a tuple from two lists
capitals = ['Washington DC', 'London', 'Tokyo']
countries = ['USA', 'UK', 'Japan']
countries_capitals = zip(countries,capitals)
print(countries_capitals)
print(*countries_capitals)
```

### Nested list comprehension

This does get a bit convoluted, so be careful. 

> Code readability should be higher priority over showing off ninja skills.  

Good for N dimensional matrices.

```python
# nested list comprehension 
# https://www.geeksforgeeks.org/nested-list-comprehensions-in-python/
ROWS = 5
COLS = 6
# range doesn't include the last count
matrix = [[c for c in range(0, COLS)] 
              for r in range(0, ROWS)]
print(matrix)

# some function of col + row
matrix_r_plus_c = [[r*COLS + c for c in range(0, COLS)] 
                                  for r in range(0, ROWS)]
print(matrix_r_plus_c)
```

Let us flatten the multi dimensional matrix to a single dimensional list.

```python
# flatten the matrix
# https://www.geeksforgeeks.org/nested-list-comprehensions-in-python/
matrix_1d = [col for row in matrix for col in row ]
print(matrix_1d)
```

How about splitting the input into different lists? Ok, let's give it a try.

```python
# nested lists - return two lists
sq_evens,cb_odds = [],[]
[sq_evens.append(x**2) if x%2 == 0 else cb_odds.append(x**3) for x in integers ]
# [f(x) if x_meets_condition else ]
print(sq_evens)
print(cb_odds)
```
