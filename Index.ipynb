```python
# some constants that we will use
TEN = 10
HUNDRED = TEN * TEN
THOUSAND = TEN * HUNDRED
MILLION = THOUSAND * THOUSAND
BILLION = THOUSAND * MILLION
```

```python

LIM = HUNDRED

# initalize
integers = [0] * LIM
# print(integers)

# set values 
for i in range(LIM):
#     print(i)
    integers[i] = i

print(integers)
```

```python
even_integers = []
for i in range(len(integers)):
    if integers[i]%2 == 0:
        even_integers.append(integers[i])
print(even_integers)
```

```python
squared_even_integers = []
for i in range(len(even_integers)):
    squared_even_integers.append(even_integers[i]**2)
print(squared_even_integers)
```

```python
sum_of_evens = 0
for i in range(len(squared_even_integers)):
    sum_of_evens += squared_even_integers[i]
print(f"sum_of_evens : {sum_of_evens}")
```

```python
odd_integers = [x for x in integers if x%2 != 0]
print(odd_integers)
```

```python
squared_odd_integers =  [x**2 for x in integers if x%2 != 0]
print(squared_odd_integers)
```

```python
import functools
sum_of_odds = functools.reduce(lambda a,b: a+b, squared_odd_integers)
print(sum_of_odds)
```

```python
# square even and cube odds
square_even_cube_odd = [x**2 if x%2==0 else x**3 for x in integers]
print(square_even_cube_odd)
```
